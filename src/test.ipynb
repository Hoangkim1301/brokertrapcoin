```python
import requests
from bs4 import BeautifulSoup

def scrape_yahoo_finance_news(stock_symbol):
    # Replace 'AAPL' with the desired stock symbol
    url = f'https://finance.yahoo.com/quote/{stock_symbol}/news?p={stock_symbol}'

    # Send an HTTP request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find HTML elements containing news headlines (modify as per the website's structure)
        headline_elements = soup.find_all('h3', class_='Mb(5px)')

        # Extract and print news headlines
        for headline_element in headline_elements:
            print(headline_element.text.strip())
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

if __name__ == "__main__":
    stock_symbol = 'siemens'  # Replace with the desired stock symbol
    scrape_yahoo_finance_news(stock_symbol)
```
